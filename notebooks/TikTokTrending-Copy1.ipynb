```python
import json
import pandas as pd
trending_json = json.loads(response.text)
```

```python
parsed_data = {
    "id": [],
    "type": [],
    "cover": [],
    "title": [],
    "subTitle": [],
    "description": [],
    "link": [],
    "round": [],
    "playToken": [],
    "keyToken": [],
    "extraInfo": []
}
```

```python
for type_result in trending_json["body"]:
    for item in type_result["exploreList"]:
        for k in item["cardItem"].keys():
                val = item["cardItem"].get(k)
                #print(val)
                parsed_data.get(k).append(val)
```

```python
parsed_df = pd.DataFrame(parsed_data)
parsed_df
```

```python
extrainfo_keys = []

for info in parsed_df['extraInfo']:
    extrainfo_keys += (list(info.keys()))
extrainfo_keys = list(set(extrainfo_keys))
extrainfo_keys
```

```python
for infokey in extrainfo_keys:
    parsed_df[infokey] = parsed_df['extraInfo'].apply(lambda x: x.get(infokey))
parsed_df.head()
```

```python
music_df = parsed_df[parsed_df['type'] == 1]
music_df
```

```python
def parse_number(num: str) -> int:
    videos = num.split()[0]
    number = videos[:-1]
    units = videos[-1]
    units_mapper = {
        "K": 1,
        "M": 1000,
        "B": 1000000
    }
    return (float(number)) * float(units_mapper.get(units))
```

```python
df = pd.DataFrame(music_df['subTitle'])
df
```

```python
df['videos_count (in thousands)'] = df['subTitle'].apply(parse_number)
df
```

```python
music_merged = pd.merge(music_df, df).drop(columns=
                                           ['subTitle',
                                            'cover', 
                                            'round', 
                                            'keyToken', 
                                            'playToken', 
                                            'extraInfo', 
                                            'type', 
                                            'extraInfo',
                                            'roomId',
                                            'fans',
                                            'verified',
                                            'relation',
                                            'challengeId',
                                            'musicId',
                                            'posts',
                                            'likes',
                                            'secUid',
                                            'userId',
                                            'challengeName',
                                            'views',
                                            'playUrl'])
music_merged
```

```python
music_clean = music_merged.rename(columns={
    'id': 'Song ID',
    'title': 'Song Name',
    'description': 'Artist',
    'link': 'Link to Trend Page',
    'videos_count (in thousands)': 'Number of Videos with Song (in Thousands)'
})
music_clean
```

```python
music_clean.to_csv('music_data0609.csv')
```
